## Applying the test- train split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_regressor = LinearRegression(fit_intercept=False)
lin_regressor.fit(X_train, y_train)
```

```python
y_pred_test = lin_regressor.predict(X_test)
y_pred_train = lin_regressor.predict(X_train)

from sklearn import metrics
rms_test = np.sqrt(metrics.mean_squared_error(y_test, y_pred_test))
rms_train = np.sqrt(metrics.mean_squared_error(y_train, y_pred_train))

print('The RMSE of the training set is: ' + rms_train.astype(str))
print('The RMSE of the test set is: ' + rms_test.astype(str)) 
```

### Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
dt_regressor = DecisionTreeRegressor()
dt_regressor.fit(X_train, y_train)
```

```python
y_pred_test = dt_regressor.predict(X_test)
y_pred_train = dt_regressor.predict(X_train)

from sklearn import metrics
rms_test = np.sqrt(metrics.mean_squared_error(y_test, y_pred_test))
rms_train = np.sqrt(metrics.mean_squared_error(y_train, y_pred_train))

print('The RMSE of the training set is: ' + rms_train.astype(str))
print('The RMSE of the test set is: ' + rms_test.astype(str)) 
```

### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor(n_estimators=30)
rf_regressor.fit(X_train, y_train)
```

```python
y_pred_test = rf_regressor.predict(X_test)
y_pred_train = rf_regressor.predict(X_train)

from sklearn import metrics
rms_test = np.sqrt(metrics.mean_squared_error(y_test, y_pred_test))
rms_train = np.sqrt(metrics.mean_squared_error(y_train, y_pred_train))

print('The RMSE of the training set is: ' + rms_train.astype(str))
print('The RMSE of the test set is: ' + rms_test.astype(str)) 
```
